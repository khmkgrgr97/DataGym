```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree  import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
import sklearn
from sklearn.preprocessing import scale
from sklearn.utils import shuffle
from sklearn.model_selection import validation_curve
from sklearn.model_selection import KFold
```

```python
df = pd.read_csv('hw3_train.csv', sep=',') 
```

```python
#len(df[df['use_short_deposits']!=0])
df = df.drop('use_savings',axis = 1)
df = df.drop('use_guarantees',axis = 1)
df = df.drop('use_derivada_account',axis = 1)
df = df.drop('use_short_deposits',axis = 1)
```

```python
X = scale(df[df.columns[1:-1]])
y = df["use_direct_debit"]
```

```python
(X_train, X_test,
 y_train, y_test) = train_test_split(X, y, 
                                     test_size=0.3, 
                                     random_state=0)
```

```python
help(DecisionTreeClassifier)
```

```python
best_score = 0
best_i = 0
for i in range(1,60,2):
    
    knn = DecisionTreeClassifier(max_depth = i)
    knn.fit(X_train, y_train)
    pred = knn.predict(X_test)
    if sklearn.metrics.f1_score(y_test, pred) > best_score:
        best_score = sklearn.metrics.f1_score(y_test, pred)
        best_i = i
    print(sklearn.metrics.f1_score(y_test, pred))
print (best_score, best_i)
```
